```python
for i in df.columns:
    if df[i].dtype == object:
        if len(df[i].unique()) < 800:
            print("The unique values present in", i ,"are: \n\n{}".
                  format(df[i].unique()),'\n')
# Note There are 5 major neighbourhood groups and many neighbourhoods in NYC and there are 3 room types in those AirBNBs
```

```python
labels = df.neighbourhood_group.value_counts().index
sizes = df.neighbourhood_group.value_counts()
plt.figure(figsize = (5,5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%',colors = ['green','red','blue','orange', 'yellow'])
plt.title('Percentage of AirBNBs group wise',color = 'black',fontsize = 15);
plt.show()
```

```python
sns.countplot(x = df["room_type"], palette = "summer")
plt.xlabel("Room Type", fontsize = 12)
plt.ylabel("Count", fontsize = 12)
plt.show()
```

```python
sns.barplot(x = df["room_type"], y = df["price"], palette = "summer")
plt.xlabel("Room Type", fontsize = 12)
plt.ylabel("Price", fontsize = 12)
plt.show()
```

```python
data_price_95 = df[df['price'] <=df.quantile(0.95)['price']]
plt.scatter(data_price_95['longitude'], data_price_95['latitude'], cmap=plt.get_cmap('rainbow'), alpha=0.4)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.show()
```
